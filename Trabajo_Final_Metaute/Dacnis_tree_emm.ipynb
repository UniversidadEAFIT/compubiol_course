## Introducción

La familia Thraupidae, comunmente conocida como las tángaras , son un grupo diverso de Passeriformes neotropical, con un amplio rango en morfología, comportamiento, patrones en el plumaje, colores, ecología y dieta, que genera un poco de confusión y problema a la hora de intentar agrupar alguna de sus especies dentro de un rango taxonómico dentro de la familia(Burns et al., 1997). La clasificación tradicional de las tángaras ha estado respaldada y estuvo permeando por  años, por la clasificación realizada por Storer (1970), en el que se describe un total de 242 especies, agrupadas en 57 géneros; al ser tradicional, las categorías taxonómicas fueron dadas basadas principalmente en morfología y comportamiento pero como se ha visto no solo en trabajos de relaciones evolutivas o taxonómicas en aves si no en diferentes grupos bióticos, es necesario de un respaldo molecular para tener una mayor seguridad de la posición y relación de las especies, ya que se ha encontrado en los últimos años que algunas de las especies que se creían que estaban relacionadas posiblemente tengan una mayor relación con otros grupos taxonómicos(Burns et al.,2016).

El intento de resolver y tratar de encontrar el mejor acercamiento de las relaciones evolutivas reales que ha ocurrido a través del tiempo, ha hecho que se estén haciendo diferentes análisis filogenéticos con diferentes metodologías y diferentes marcadores moleculares, que puedan dar la respuesta más cercana a la realidad y que puedan dar cuenta de las relaciones filogenéticas de los grupos, que a parte de dar información sobre los grupos monofiléticos existentes, dicha información ayude al entendimiento de la biogeografía, evolución, comportamiento, ecología para poder tomar acciones de conservación si son necesarias.



## Metodología

Para la reconstrucción filogenética, se buscaron las secuencias de las especies del género *Dacnis* en el buscador Taxonomy del NCBI, en donde se descargaron un total de 31 secuencias: 8 para COI, 14 para CYTB y 9 para ND2. Se escogieron estos marcadores, debido a que eran los genes que tenían mayor representación en la base de datos para ese género. Como grupo externo se escogió la especie *Ramphocelus passerinii*, descargando una secuencia para cada uno de los marcadores escogidos.

Posteriormente, se realizó el alineamiento de las secuencias por medio de MAFFT. Los resultados del alineamiento fueron sometidos a JModelTest, para determinar cuales eran los mejores modelos evolutivos que soportaran mis datos. Luego, para la construcción de las topologías se utilizó GARLI (2.01) con 10 repeticiones para la búsqueda del mejor árbol para cada uno de los marcadores y 100 réplicas de Boostrap. Finalmente, mediante Summtrees se sobreponieron los mejores árboles encontrados por GARLI y los valores de boostrap, que posteriormente fueron graficados en Biopython mediante el paquete Phylo.  

## Resultados

#### *Modelo evolutivo*

Según los valores de BCI (Bayesian Information Criterion ) para los marcadores genéticos COI y ND2 el mejor modelo evolutivo fue el KImura-2-parameter K80+I el cual tiene en cuenta dos parámetros: la probabilidad de substituciones nucleotídicas por transiciones o por transversiones, siendo más comunes las primeras, en las que puede ocurrir transiciones entre A<-> G o C<->T; y las transversiones entendidas como el cambio de una pirimidina a una purina o viceversa. 

Para el marcador CYTB, el mejor modelo evolutivo fue el Symetrical Model SYM el cual asume que cada base tiene la misma frecuencia y trabaja con una matrix simétrica de substituciones en donde la probabilidad de pasar de A a T en la misma de T a A.

## Discusión

Dado los resultados obtenidos al realizar la reconstrucción filogenética del género *Dacnis* , se podría decir que el mejor marcador para estos datos, es el mitochondrially encoded NADH dehydrogenase 2 ND2, que junto con el modelo evolutivo K80, se pudo obtener una topología con ramas en su mayoría con buenos soportes >75, agrupando monofileticamente las especies y dejando por fuera la especie *Ramphocelus passerinii* como debe ser, ya que es un grupo externo que ayuda a darle una raíz al árbol. Las otras topologías, además de que tienen valores de soportes muy bajos, agrupó a *R. passerinii* con algunas especies de *Dacnis*, mostrando que no se debe de confiar en este árbol, ya que no está mostrando realmente las relaciones evolutivas entre el grupo. 

Al buscar literatura sobre filogenias de la familia thraupide, se logró encontrar una topología del género *Dacnis* en donde también utilizaron como marcador molecular el ND2. Al comparar la topología dada por Burns et al., 2014 se encuentra un comportamiento muy parecido al árbol obtenido en este trabajo, mostrando que *Dacnis cayana* y *Dacnis flaviventer* en efecto son taxa hermanas y en donde las demás especies también concuerdan con las relaciones evolutivas. La filogenia encontrada, ayuda a terminar de confirmar que el mejor marcador molecular para hacer una reconstrucción filogenética de este grupo es el ND2. Sin enmbargo, es de anotar que para la construcción de estos árboles, se debió tener todas las especies del género y un poco más de secuencias para cada uno de los marcadores, para poder tener un resultado más certero y confiable.
