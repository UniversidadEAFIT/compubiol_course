**2. Introducción**
*Ricinus Communis* es una planta oleaginosa de la familia Euphorbiacea[1] y en la actualidad se considera una de las especies más importantes del reino vegetal y se cree que a partir de los diferentes componentes obtenidos de su semilla se pueden adquirir cerca de 700 productos industriales, entre los que se encuentran medicamentos, cosméticos, lubricantes y su uso como biodiesel[2].

*Ricinus Communis* produce en mayor cantidad el ácido ricinoleicoo, el cual, basados en el conocimiento actual es sintetizado en el plástido, donde se exporta al citoplasma y tras un activición mediada por la Coenzima A es incorporado en el retículo endoplasmático, donde sufre una serie de hidroxilaciones y acilaciones determinantes para el resultado final[3,4].

KEGG (Kyoto Encyclopedia of Genes and Genomes) es un recurso de base de datos para comprender funciones y utilidades del sistema biológico, como la célula, el organismo y el ecosistema, a partir de información de nivel molecular, especial conjuntos de datos moleculares generados por secuenciación de genoma, y el lenguaje KGML es un formato de intercambio de rutas KEGG, que permite el trazado automático de las rutas KEGG y proporciona intalaciones para realizar análisis computaciones y modelado de redes de genes, proteínas y redes químicas[5].

El objetivo principal de este trabajo consiste en realizar una análisis de las rutas metabolicas involucradas en la sintesis y enlongación de ácidos grasos en *Ricinus Communis*, facilitando su observación, reconociendo la disponibilidad de información existente para este organismo y este notebook se presenta como una introducción a los módulos de Bio.KEGG y Bio.Graphics.KGML, con ejemplos e incluyendo manipulacion de datos.

**5. Análisis, discusión y Conclusiones**
