```python
def sumOddNumbers(a,b):
    array=list(range(a,b+1))
    suma=0
    i=0
    largo=len(array)
    while i<largo:
        if (array[i]%2 != 0):
            suma=suma+array[i]
        i=i+1
    return(suma)
sumOddNumbers(4118, 8554)             
```
