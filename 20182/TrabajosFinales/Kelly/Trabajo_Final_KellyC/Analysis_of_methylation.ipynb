#### Abstract

DNA methylation is a pervasive feature of bacterial genomes and an important epigenetic layer that is fundamental for microbiological processes, microbe adaptability and disease pathogenicity. Complete DNA methylomes for several organisms are being analyze, helping clarify the evolutionary story of this epigenetic mark and its distribution in key genomic elements [4]. This computational work analyzes the genome of its methylome. The results show that methylation takes place in the specificity 20-TAAAm6-20 and that the CCGG site possible MTase recognition sequence because is the only shared region in most of the methylation context sequences.


#### Introduction

Scientists have sequenced thousands of bacterial genomes but now they are focusing on sequencing data from another layer of information that appears to be critical for understanding where, when, and how normal functions of bacteria cells are deviate from normal bacteria behavior: epigenetic modifications. Epigenetic phenomena include prions, genomic imprinting, histone modification and DNA methylation patterns, were each of these affects gene expression [1]. DNA methylation patterns was originally discovered in the context of restriction-modification (R-M) systems, in which a restriction enzyme (endonuclease) recognizes and cuts a specific region of the DNA as a form of cellular defense, unless that region has been methylated by a cognate DNA methyltransferase [1, 6].

We have known the role of some eukaryote components of R-M systems like 5-methylcytosine (m5C), that is also found in bacteria. In addition, two more modified bases, N6-methyl-adenine (m6A) and N4-methyl-cytosine (m4C), are common in bacterial genome, where they function as a recognition place of DNA motifs by DNA-binding proteins like DNA methyltransferase (MTases) (DNA methylation in bacteria: from the methyl group to the methylome).  Also, these MTases serve regulatory roles as with the Dam MTase of *E. coli*, which introduces methylations m6A base that play a key role in DNA repair and have important effects during the initiation of replication, are implicated in regulating gene expression, phase variation and pathogenicity [3, 4, 5]. 

Up to the present, there has been no genome-wide assessment of the extent of DNA methylation by known MTases such as *E. coli* Dam MTase. It is not known if their methylation specificities are as precise as the customary recognition sequences suggest or whether the enzymes methylated in different recognition sequences motifs. On top, there has not been a great deal of research into the methylation pattern of the majority bacterial genomes. Hence, in order to find the methylated pattern and to possible assign recognition sequences for the MTase, int this computational approach it will be used the Bio.motifs package to perform motif analysis in a *B. subtilis* genome [2, 4].

#### Methodology
1. Data was acquire from PacBio Methylation Context Data. 
2. It was made a grep for CCGG across *B. subtilis* genome.
3. The sequences were exported in a fasta format.
4. Motif object (list of unaligned CCGG) was created.
5. A sequence logo was created to identify the methylation specificity of DNA MTases.
6. A multiple sequence alignment of CCGG was performed using MAFFT software.
7. A sequence logo was created to identify the recognition sequence of DNA MTases.

#### Results

```python
#Required libraries to make an screening and target sequences motifs of methylation and recognition sequence.
from Bio.Seq import Seq
from Bio import motifs 
```

```python
with open("CCGG.txt") as meth_context_seqs:
    content = meth_context_seqs.readlines()
content = [x.strip() for x in content] 
print(content)
```

```python
#Creation of a motif object with the methylation context sequences which is saved in the form of list.
m = motifs.create(content)
print(m)
```

```python
#The length of the motif is defined as the sequence length, which should be the same for all instances, in this case is 41.
len(m)
```

```python
#It can be known the counts of each nucleotide at each position printing a .counts matrix
print(m.counts)
```

```python
#From the largest value in the corresponding columns of the .counts matrix are obtain the associated consensus sequence of methylation of the motif 
m.consensus
```

```python
#As well, the smallest values in the columns of the .counts matrix are obtain:
m.anticonsensus
```

```python
#In addition it can be obtain sequence in which ambiguous nucleotides are used for positions where there are multiple nucleotides with high counts.
#N follows the IUPAC nucleotide ambiguity codes: N is either A, T, C, or G.
m.degenerate_consensus
```

As you can see, the  recurring patterns in the motif sequence is TAAA y CCGGT

#### Discussion and conclusion
Motif analysis resulted in an approximate identification of the methylation specificity of DNA methyltransferases (MTases), from the genomic positions detected as methylated, and the MTase recognition sequence of *B.subtilis* genome. Methylation takes place as indicated in Graph1 in the specificity 20-TAAAm6-20, where the A in the position 21 is the most methylated base of the methylation specificity. This makes sense, since this base is the only one that is present for each sequence analyzed. In the other hand, after alignment it can be observed in Graph2 that CCGG site is a possible MTase recognition sequence, which is the region where the enzyme is coupled to perform subsequent methylation. Since the recognition sequence CCGG is not well conserved throughout all the sequences, it is required further characterization to assert this. 

Despite this analysis have relatively little information about methylation specificity and recognition sequence of MTases of *B. subtilis* genome, this a valuable computational approach that should become
much easier in the future to yield more suitable and accurate data in newly sequenced genomes. It is fundamental to keep analyzing methylation patterns for understanding fundamental microbiological processes,
microbe adaptability, disease pathogenicity and how the addittion or removal of methylation patterns may fundamentally change the biology of the organism under study.
