### Planteamiento del problema

Alrededor de 120 países tropicales en los 5 continentes cultivan banano y hoy en día los cultivares que se comercializan son clones sin semillas, provenientes del genoma A (de la especie Musa acuminata) o de la unión de los dos genomas A y B (de las especies Musa acuminata y Musa balbisiana); y dentro de esos cultivares comerciales encontramos algunos como “Grand Naine”, “Poyo”, “Williams” (Bakry, et al., 2009), “Valery” (Marín, et al., 2003), entre otros. 

Ya que la reproducción de estos cultivares es de manera asexual vía micropropagación vegetativa o retoños, su base genética se reduce poco a poco y esto conlleva a la producción de cultivos poco resistentes a plagas y enfermedades (Miller, et al., 2010); las cuales, junto con los efectos del cambio climático que limitan y retrasan toda la cadena productiva y la aplicación extensiva de agroquímicos, son las principales amenazas que reducen la producción y pueden llegar a acabar con los cultivos de banano (El Informador , 2017). Por tanto, es de vital importancia mejorar la producción y tener alternativas de desarrollo sostenible en el sector bananero mundial, en especial en aquellos países que proveen el producto (Ecuador, Filipinas, Costa Rica, Colombia y Guatemala) a los grandes compradores (Estados Unidos y la Unión Europea) (FAO, 2018).

En Colombia los cultivos de los cultivares comerciales de banano se ven amenazados constantemente por diversas enfermedades y plagas que pueden generar pérdidas por  afectación directa de la planta o por los altos costos de manejo de “erradicación de focos” y el tiempo en que la tierra permanece infertil e improductiva (DANE, 2016). Algunos de los agentes microbianos que pueden causar enfermedades son los hongos Mycosphaerella fijiensis y Fusarium oxysporum y la bacteria Ralstonia solanacearum; en cuanto a plagas, los cucarrones Cosmopolites sordidus, los nemátodos Radopholus similis,  Pratylenchus spp. y Meloidogyne incognita y M. javanica, y las moscas blancas Aleurodicus floccissimus, son quienes causan más estragos en los cultivos (DANE, 2016). 

Debido a lo anterior, se buscan alternativas seguras y efectivas para controlar y evitar al máximo, los daños en cultivos por enfermedades y plagas. Por ende, puede ser un punto clave para plantear nuevas alternativas de control, estudiar la condición ancestral de estos cultivares comerciales; como por ejemplo variedades silvestres como la  diploide nativa Calcutta 4, originaria del Sudeste Asiático, la cual ha demostrado ser resistente a diferentes ataques fúngicos y bacterianos, así como a nemátodos (Miller, et al., 2010). Para ello es importante conocer la diversidad y riqueza de la microbiota, el core tanto de ésta como del microbioma, entender la funcionalidad de cada grupo de microorganismos encontrados para combinar análisis moleculares (metagenómica y metatrancriptómica) que logren dilucidar la importancia de la microbiota en la planta, ya que su complejidad puede ser consecuencia del reclutamiento activo de microorganismos por parte de la planta y la transferencia oportunista de microorganismos no patógenos (Vandenkoornhuyse, et al., 2015).

Finalmente, la pregunta que se plantea para este proyecto es si la resistencia a patógenos de una variedad ancestral como Calcutta 4, se debe a una microbiota endófita nativa distintiva respecto a los cultivares comerciales y a diferencias en abundancia de las comunidades bacterianas presentes en ambas plántulas; por lo tanto, partimos desde la hipótesis de que en efecto, existe una microbiota endófita diferencial entre Calcutta 4 y Williams, destacándose una mayor riqueza de microorganismos y un mayor equilibrio biológico para Calcutta 4 en comparación con el cultivar Williams. Estas son las preguntas base para en el futuro desarrollar un suplemento biológico, basado en aquellos microorganismos presentes en Calcutta 4 que no se encuentran en Williams; y poder proporcionar a las plantas susceptibles a los patógenos, una microbiota que podría ser la causal de la “resistencia” de  una variedad ancestral a las enfermedades mencionadas. 






### Metodología: aislamiento de microorganismos endófitos nativos

Partimos de  plantas in vitro de Calcutta 4 y Williams germinadas a partir de semillas y a partir de meristemos de 9 meses de desarrollo provistos por Meristemos de Colombia, respectivamente. Plantas de invernadero Calcutta 4 de un sistema natural en etapas iniciales de desarrollo (2 meses) en suelo de la finca Montelindo de la Universidad de Caldas en Santágueda, Caldas; y plantas Williams de 4 meses de desarrollo en invernadero.

Con el fin de determinar la microbiota endófita nativa que compone los diferentes órganos de las plantas de banano Calcutta 4 y Williams, se desinfectará y verificará la desinfección de hojas, pseudotallo y raíz de las plantas para verificar la no presencia de organismos epífitos. Se macerarán con Ultraturrax para realizar diluciones seriadas que se sembrarán en los medios AIA para el aislamiento de Actinomycetes (0,1 g/L de L- Asparagina; 0,5 g/L de K2HPO4; 0,001 g/L de FeSO4*7H2O; 2 g/L de caseinato de sodio; 0,1 g/L de MgSO4*7H2O y 15 g/L de Agar) (Chaudhary et al., 2013), PIA para el aislamiento de Pseudomonas (1,4 g/L de MgCl2; 20 g/L de proteosa peptona; 10 g/L de sulfato de potasio; 0,025 g/L de triclosán y 13,6 g/L de bactoagar (281210, Difco) a pH 7,0) (Keeven y de Cicco, 1989), YDC para aislar bacterias Gram negativas (20 g/L de CaCO3, 20 g/L de glucosa, 10 g/L de extracto de levadura y 15 g/L de bactoagar a pH 7,0) (Atlas, 2010), TSA 50% medio enriquecido de amplio espectro (9 g/L de bactoagar BD y 20 g/L de TSA 105458, Merck) (Atlas, 2010) y R2A medio enriquecido (0,5 g/L de proteosa peptona, 0,5 g/L de casaminoácidos, 0,5 g/L de extracto de levadura, 0,5 g/L de glucosa, 0,5 g/L de almidón soluble, 0,3 g/L de K2HPO4, 0,05 g/L de sulfato de magnesio, 0,3 g/L de piruvato de sodio y 15 g/L de bactoagar a pH 7,0) realizando dos repeticiones por trés réplicas biológicas por 4 cajas por dilución. 

Estos medios se seleccionan con el fin de aislar microorganismos de forma diferencial respecto a dos medios donde se aíslan poblaciones totales. Los morfotipos obtenidos se purificarán y conservarán en TSB + glicerol 20% a -80ºC, para posteriormente realizar la extracción del DNA de cada aislado por choque térmico y/o por Kit (Ultraclean Microbial DNA). El DNA se amplificará por PCR (Mix: 0,02 U/uL de Taq Polimerasa 5 U/mL; 0,2 uM de primers (8F y 1492R); 0,2 mM de dNTP's-Mix; 1,5 mM de MgCl2; 1X de buffer de la Taq, 1 uL de DNA y agua ultrapura para 50 uL) así: desnaturalización a 94°C (3 min), 30 ciclos de amplificación, desnaturalización a 94°C (30 s), alineamiento 60ºC (20 s), extensión a 72°C (90 s) y extensión final a 72°C (5 min). Los productos de la PCR se evaluarán en NanoDrop 2000 y en agarosa al 1,0 %. 

Las muestras se secuenciarán por método Sanger y los resultados se evaluarán empleando un PIPELINE bioinformático y posteriormente, se ingresarán al BLAST del NCBI tales secuencias con el fin de identificar los microorganismos aislados.



### Metodología: desarrollo del Pipeline

Tras obtener los resultados de la secuenciación por método Sanger, se recibieron 3 carpetas, cada una con 4 archivos por cada cepa secuenciada, de extensiones .ab1, .pdf, .phd.1 y .txt

Para proceder con la identificación de cada una de las cepas (206 en total):

Desde la terminal usando el comando cd desplazarse a la carpeta finalCompu, llame python3, el script realizado, la carpeta datos, el n deseado y el score: se recorrieron las 3 carpetas para extraer únicamente los archivos .ab1 de los que se obtendrían las secuencias de ADN para posteriormente, realizar el trimaje de las mismas usando una ventana de 6 bases y un score de 30 (estos valores podrán ser modificados por quien use el código).

Al finalizar la extracción de los archivos y su respectivo trimaje, estos nuevos archivos generados serán de extensión .txt y se guardarán en una carpeta independiente "resultados"*. Estos archivos finales se podrán cargar en la página del BLAST y se eligirá el organismo que posea un genoma completo y el mayor 'Total Score' (estos parámetros son de libre elección según el usuario).

*NOTA: para esta versión, la carpeta "resultados" deberá existir previamente en la carpeta donde se encuentren los datos y el script.

### Conclusiones

En muchas ocasiones, la secuenciación por el método de Sanger no es la más precisa y ésto se puede evidenciar tanto con el cromatograma enviado por la empresa, como con la secuencia trimada obtenida a través de este pipeline.

A la hora de realizar un trimaje, no se puede dar por sentado que en el medio de la secuencia se obtendran los valores Phred más altos pues se debe considerar que:

1. El equipo se puede tardar un poco más en calibrarse y por tanto, los picos definidos pueden observarse al cabo de tiempos variables. Lo mismo sucede con el final, los reactivos pueden no siempre "acabarse" o perder afinidad en el mismo tiempo.

2. Hay muestras que pueden estar contaminadas y por esta razón, ni siquiera en la mitad, se logran encontrar valores de Phred mayores o iguales al score deseado; lo cual terminará arrojando trimajes de 10, 2 e incluso 1 nucleótido(s).

Por lo tanto, es importante saber que un trimaje siempre va a depender del score indicado y del tamaño de la ventana elegida y que en muchas ocasiones, no es útil utilizar la secuencia que se obtiene al final de un trimaje ya que puede ser inferior a 10 nucleótidos y esta cantidad no es suficiente para realizar la identificación del organismo (hay que recurrir a la secuencia completa).
