# Introduction

The Araceae, also known as the aroid family, comprise about 118 genera and 3,800 species of monocotyledonous herbs and vines (Henriquez *et al*., 2014). Although the family has natural presence in every continent except Antarctica, it is predominantly tropical: most aroid genera occur in tropical Asia and the New World tropics (Grayum, 1990). While the Araceae are exceptionally diverse by nearly all phenetic criteria (e.g roots, leafs, flowers), present evidence suggests that the family is a natural and monophyletic group (Grayum, 1990).

In terms of morphology, the aroid family may be characterized by the following probable synapomorphies (Grayum, 1984, 1987): a bractless, spicate inflorescence; extrorse anther dehiscence; a unipistillate gynoecium; a berrylike fruit; amoeboid anther tapetum; helobial (i.e., haustorial) endosperm development; and the (probably) universal presence of tannins. To this list may be added the following features which, though presumably plesiomorphic, help characterize the family: bifacial leaves; presence of raphides; lack of ethereal oil cells; hypogynous, actinomorphic flowers; basifixed anthers; successive pollen mother-cell cytokinesis; absence of perisperm; absence of silica; and absence of tricin. All of the above characteristics are universally present, as far as is known, in all genera of Araceae; furthermore, no other monocot taxon possesses them all in combination (Grayum, 1984, 1987).

As in most plant families, phylogenetic relationships have been analyzed through morphology. This is not necessarily wrong, but it can be imprecise; and through recent sequencing technologies, relationships between taxa can be thoroughly -and more precisely- elucidated. Nonetheless, not all plant groups have received the same attention, and the only study regarding phylogenomics of the Araceae was recently published by Henriquez *et al* (2014).

To the best of my knowledge, no work regarding phylogenomics of the genus *Monstera* has been done before. Even though since Humboldt's time *Monstera* has been recognized as one of the most common genera in the Neotropics (Botting, 1973), the relationships between its species have relied entirely in morphology. This project intends to analyze published nucleotide data in the NCBI to try to understand *Monstera* phylogenetic relationships through DNA sequences.

# Methods

Using the Taxonomy Browser of the NCBI, 66 nucleotide records of various *Monstera* species were downloaded. However, not all the sequences corresponded to vouchered specimens, which reduced the number of records to 63. Moreover, taking into account the minimum number of records to reconstruct phylogenetic relationships, the number of records were further reduced to 50, corresponding to three genetic markers: matK, rbcL and trnL-trnF; with 19, 23 and 8 records, respectively. 

Then, the records corresponding to each genetic marker were aligned using MAFFT (7). Using JModelTest (2.1.10), the best evolutionary model was determined using the Bayesian Information Criterion (BIC). Later on, using GARLI (2.01) maximum likelihood trees were reconstructed, 2 for each genetic marker: 100 search reps. (which results in the best tree) and 1000 Bootstrap replicates to validate the best tree. Finally, the resulting trees were both imported to Fig Tree (1.4.4) and converted to .newick (Phylo.draw) in order to visualize the results.

# Results

The best evolutionary model for the three markers corresponded to K80, which distinguishes between transitions (A->G, i.e. from purine to purine, or C->T, i.e. from pyrimidine to pyrimidine) and transversions (from purine to pyrimidine or vice versa). The K80 model assumes that all of the bases are equally frequent (πA=πG=πT=πC=0.25).

# Discussion

In first place, it is interesting that the best evolutionary model for the three markers is K80. In fact, with both Akaike Information Criterion (AIC) and Bayesian Information Criterion (BIC) the resulting model is the same, thus confirming that the output is not dependent -in this case- of the used criterion. This means that, for the *Monstera* species, all the nucleotides have approximately the same frequency.

On the other hand, it is quite easily detected that the trees presented through Fig Tree are better presented. While the Phylo.draw tool allows for a quick visualization (without having to install any other program), the BS values, branch lengths and tip names are not easily visualized. Fig Tree does a better job in terms of presenting the information.

As mentioned before, three markers were used: (1) matK (Maturase K), a plant plastidial gene (Zoschke *et al*., 2010), which encodes a protein that splices Group II introns (Ahlert *et al*., 2006); (2) rbcL , also a plastidial gene which encodes to the large subunit of the ribulose-bisphosphate carboxylase (also known as RuBisCO), an essential protein in the Calvin Cycle, and thus for photosynthesis itself (Cooper, 2000); (3) trnL-trnF, which consists of the trnL gene, a group I intron, and the trnL-F intergenic spacer (Huang *et al*., 2009).

Now, in terms of the reconstruction, the trees are not particularly good, which is a result itself. Using matK, while there are good BS values, the species being grouped are not monophyletic since *M. deliciosa* and *M. tenuis* are intermixed. However, there is a small cluster of three species with 100 BS value with a long branch, which corresponds to species not being formally described (as noted by the sp.), this could mean that they represent one unique species. Nonetheless, recognizing that the other species are intermixed, this is a bold and imprecise guess. On the other hand, using rbcL the results are worse. There is an absolute mess regarding monophyly, *M. deliciosa* being intermixed with both *M. adansonii* and *M. tenuis*. In fact, this phylogeny could be considered completely useless, since it doesn't at all. Finally, using trnL-trnF noncoding region, the results are not any better, there is not any clear monophyly inside the genus and the species are completely intermixed. 

However, this is an interesting result which confirms, at least in part, why there has not been any phylogenomic approach to *Monstera*. With the sequences readily available via the NCBI the results are ambiguous. Moreover, as the Barcode of Life Database (BOLD) reports, there is not an universal marker that applies to plants; while there are some markers that help and apply to particular groups (e.g Orchids or Palms), it is rarely maintained in higher taxonomic groups such as families or orders. In fact, matK is the most used marker among plants (Jing *et al*., 2011), and this is an example in which it does not help much in elucidating relationships among a genus. 

In this sense, a bigger study must be done in order to understand the genetic relationships in *Monstera*, and in fact in all the Araceae family. While there is a phylogenomic approach to the family, even the authors mention that further analysis must be made in order to understand the relationships among the genera (Henriquez *et al*., 2014).

# Conclusions

This project intended to use the readily available nucleotide sequences of the NCBI to reconstruct phylogenetic relationships within the *Monstera* genus. While the markers used (matK, rbcL, trnL-trnF) were not useful, this is a result itself. Ambiguity among the genus using a genetic approach cannot be solved with the sequences and markers used.

More studies regarding the phylogenomic relationships (such as determining a good marker for the Araceae family) must be done in first place. In fact, this type of studies are urgently needed to change the traditional taxonomy of plants, which is based almost completely in morphology, and turn it into a genomic approach; or even better, an integrated assessment of both morphology and molecular data.
