### List-1 > first_last6

```python
def first_last6(nums):
    if (nums[0] == 6) or (nums[len(nums)-1] == 6:
        return True
    else:
        return False
```

### List-1 > common_end

```python
def common_end(a, b):
    if (a[0] == b[0]) or (a[len](a)-1 == b[len(b)-1]):
        return True
    else:
        return False
```
