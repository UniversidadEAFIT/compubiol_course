Ejercicios String-1

```python
#hello_name
def hello_name(Oscar):
    return "Hello " + Oscar + "!";
  
#make_out_word
def make_out_word(Oscar, Santi):
    return Oscar[0] + Oscar[1] + Santi + Oscar[2] + Oscar[3]
  
#first_half
def first_half(str):
    return str[: len(str)/2]
  

#Ejercicios List-1

#first_last6
def first_last6(nums):
    return nums[0] == 6 or nums[-1] == 6
  
#max_end3
def max_end3(nums):
    return [nums[0]]*3 if nums[0] >= nums[-1] else [nums[-1]]*3 
  
#reverse3
def reverse3(nums):
    return [nums[2],nums[1],nums[0]]
```

```python
#Ejercicio Warmup-1 

#near_hundred
def near_hundred(n):
    return ((abs(100 - n) <= 10) or (abs(200 - n) <= 10))
  
  
#Ejercicio Logic-1

#love6
def love6(a, b):
    return (a==6 or b==6) or (a+b==6 or abs(a-b)==6)
```
