## CodingBat List

```python
def first_last6(nums):
    if nums[0] == 6 or nums[(len(nums)-1)]==6:
        return True
    else:  
        return False
```

```python
first_last6(["a",1,2,5])
```

## CodingBat Logic:

```python
def cigar_party(cigars, is_weekend):
    if ( 40 <= cigars <= 60 ) and is_weekend == False:
        return True
    elif ( cigars >= 40 ) and is_weekend == True: 
        return True
    else:
        return False
```

```python
cigar_party(72,True)
```
