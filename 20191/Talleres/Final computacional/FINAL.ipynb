“Alineamiento de secuencias entre Bacillus amyloliquefaciens y Bacillus pumilus en su región productora de sintetasas peptídicas no ribosomales”

Santiago Mazo Zapata


INTRODUCCIÓN

En la actualidad existen un sin fin de organismos patógenos, entre ellos se encuentran los fitopatógenos y uno de los que toca a Colombia y hace perder gran cantidad de dinero en algunas de las esferas de la economía nacional es Colletotrichum Acutatum, Colletotrichum, es un género de hongo que hace parte de la familia Glomerellaceae, normalmente, estos son simbiontes en plantas a manera de endófitos, sin embargo muchas de las especies se caracterizan por ser fitopatógenos y causar enfermedades graves en partes de la planta y sus frutos (Rodríguez et al., 2008). Entre los patógenos conocidos se encuentra la especie acutatum, que causa la enfermedad conocida como antracnosis. Esta enfermedad en general presenta manchas hundidas de diversos colores en las hojas y necrosis en los nervios de las mismas además de generar síntomas en los tallos, frutos y flores. Generalmente la enfermedad desencadena en marchitamiento y pérdida del cultivo, lo que genera grandes pérdidas económicas a ciertos sectores de la industria como lo son la agricultura de exportación. De hecho C.acutatum según (Dean et al., 2012) se encuentra entre los 10 fitopatógenos más incidentes a nivel mundial atacando diversidad de frutos como mangos, papayas, manzanas y fresas, además de atacar ampliamente los lupinos. Teniendo en cuenta lo grave que puede resultar este fitopatógeno es normal que las industrias busquen proteger sus cultivos utilizando fungicidas. El problema es que muchos de estos pueden resultar altamente tóxicos para algunos otros factores bióticos y además a la larga los patógenos resultan adquiriendo resistencia (Whals et al., 2005). Actualmente este fitopatógeno se trata con fungicidas tales como mancozeb y clorotalonil ambos atacan al hongo desde el contacto con las esporas, y tienen de cierta forma un porcentaje bajo de que se produzca resistencia por parte del patógeno, sin embargo tienen efectos dañinos secundarios, por ejemplo el mancozeb tiene componentes carcinógenos y se puede introducir por vía inhalatoria (Runkle J et al ., 2017). Por su parte el clorotalonil tiene las ventajas anteriormente mencionadas pero también en exceso puede causar daño renal y tiene componentes igualmente carcinógenos y que pueden poner en riesgo la salud humana (Carrington, Damian., 2019). Otra forma de atacar el patógeno es a base de tratar los suelos, generalmente esto se hace con cloropicrina también conocida como nitrocloroformo. Es un compuesto químico que se utiliza actualmente como antimicrobiano de amplio espectro, fungicida, herbicida, insecticida y nematicida cuya fórmula molecular es CCl12NO2 (Desaeger et al., 2004), no obstante, y aunque ha dado buenos resultados combatiendo el hongo, el “anexo 2 del impacto Ambiental del Plan Nacional de Eliminación del Consumo de Bromuro de Metilo en México (1998-2013)” ha identificado varios problemas y desventajas al usar dicho compuesto entre las cuales están las siguientes: - Riesgo de contaminación de las aguas subterráneas; - Riesgo de transporte a larga distancia atmosférica; - Riesgo para los organismos acuáticos; - Riesgo para las aves y los mamíferos; - Potencial de riesgo oral y de contacto para las abejas de la exposición a los residuos; - Falta de datos en relación con el metabolito dicloronitrometano y las impurezas de la sustancia activa causando posible riesgo de contaminación de agua y sedimentos por exposición a la cloropicrina y el metabolito dicloronitrometano. Por lo anterior se ha venido buscando otras formas de tratar a este patógeno con métodos menos invasivos y que no tengan repercusiones tan amplias en cuanto a la resistencia a largo plazo. Entre los métodos que se han contemplado dentro del control biológico existe el uso de lipopéptidos que son metabolitos secundarios producidos por sintetasas peptídicas no ribosomales en B. subtilis que se agrupan en 3 familias, iturinas, fengicinas y surfactinas. Estos, a partir de un mecanismo que no está descrito en su totalidad, permeabilizan la membrana celular exponiendo el citoplasma y sus organelas causando un desbalance en sus maquinarias y provocando que no pueda subsistir (Hamoen, et al.,2003). A parte de los lipopétidos, existen otros compuestos derivados de algunos extractos de árboles, como los extractos de aceite del árbol del té (TTO), que se extrae de Melaleuca alternifolia, el cual es su mayor composición tiene un terpeno conocido como Terpinenol-4 y se tiene como el causante principal de la actividad anti-fúngica. Este compuesto, en conjunto con γ –terpina, α –terpina y 1,8-cineol (los otros componentes del extracto) de cierta forma no descrita en su totalidad ataca la mitocondria en organismos como por ejemplo Botrytis cinérea, afectando tanto su parte estructural, como la producción de ATP, por lo que se sugiere ataca la fosforilación oxidativa en alguna de sus etapas, haciendo que el organismo colapse en su funcionamiento y posteriormente muera con facilidad (Yonghua et al., 2017). 


Ahora entendiendo que biomoléculas pueden atacar este fitopatógeno, resulta para este caso muy interesante comparar las secuencias de B. amyloliquefaciens y B. pumilus, en cuanto a su región para la producción de lipopéptidos, su justificación es que B.pumilus causa precisamente intoxicación alimentaria y es interesante ver dos especies distintas a subtilis que es la líder en cuanto a control biológico, entonces resulta interesante conocer como pueden variar esto y a qué se puede deber. 


METODOLOGÍA

La metodología en sí misma resulta sencilla, lo primero que se hizo fue alinear dichas secuencias y una vez alineadas se procede a concatenar, para este preciso alineamiento se trabajó con las secuencias del gen BAMF_RS22005 para B. Amyloliquefaciens y UP12_RS02015 para B.pumilus  debido a que son los genes de las sintetásas peptídicas no ribosomales, las cuales a posteriori van a producir los lipopéptidos o compuestos con esta naturaleza

DISCUSIÓN

Como se puede observar entonces hay un score bastante alto que para el momento de analizar dichas secuencias se traduce en la similitud que pueden tener estas, explicando en un principio la cualidad que tienen para producir compuestos enzimáticos no ribosomales, donde se encuentran la fabricación propia de lipopéptidos, en principio entonces, parece que estas dos bacterias se asemejan mucho entre ellas y probablemente entonces se asemejen mucho con B.subtilis, por lo que probablemente ambas puedan servir también como control biológico de C.acutatum.

Cabe resaltar entonces que esto se logró mediado por la alineación de un modelo local, debido a que se quería comparar las zonas más parecidas de la secuencias, ya que si se hace de manera global y se intenta cubrir totalmente pueden haber errores en cuanto a que no se puede saber si las secuencias por su parte tienen el mismo inicio y el mismo final.

Reportado entonces en otras literaturas me gustaría mostrar y resaltar la importancia de conocer esto, pues como control biológico puede ser altamente usado, por lo que es menester conocer la mayor cantidad de variaciones de Bacillus que puedan tener similitud con estas secuencias en cuanto a la parte de síntesis peptídica no ribosomal por ejemplo en otro estudio realizado por Yabing et al. (2012), se trataba también un fitopatógeno fúngico en este caso Sclerotinia sclerotiorum, el cual por el mismo uso de fungicidas a nivel descontrolado ha provocado que ciertas poblaciones hayan adquirido resistencia, por lo que se inhibió con mezclas de compuestos como el azoxistrobina, SHAM, y thiram.

 En principio la azoxistrobina inhibió el crecimiento micelial durante las primeras 2 h, luego, durante 48 h con la mezcla de la azoxistrobina y el ácido salicilhidroxámico o SHAM, que es un conocido inhibidor de la respiración alternativa, potenció esta inhibición en cuanto al crecimiento micelial y de su respiración, haciendo que como conclusión haya la necesidad de estudios adicionales a estos compuestos, pero apunta mayormente a que tienen efectos dañinos sobre la mitocondria nuevamente. 

Siguiendo con estos tipos de compuestos hay otro estudio realizado por Fangli Wu et al. (2010) el cual haciendo uso de la propamidina, que a grandes rasgos es una diamida aromática, se llegó a la conclusión de que decrecía la actividad mitocondrial del complejo III en B. cinerea y a parte de esto destrozaba el micelio.

 El funcionamiento general que se llegó a describir fue que la propamidina se une a un surco menor de un gen que codifica una subunidad del complejo III e inhibe por consecuencia la expresión de este gen, por tanto disminuye el control respiratorio e influyó en la producción de energía, provocando posteriormente anomalías en la ultra estructura y deteniendo el crecimiento. 

Esto hace pensar que puede pasar si por ejemplo la actividad de la propamidina, con compuestos lipopéptidos como lo son la fengicina C que por su estructura podría sugerir tener mayores propiedades inhibitorias según Villegas-Escobar et al. (2013), podrían ocasionar a nivel estructural en un patógeno fúngico como los anteriores descritos, algo que llama la atención a investigar sobre la relación inhibitoria contra esta organela. 

Otro estudio relacionado con respecto a la interacción que tienen las etapas para la obtención de energía, en este caso la beta-oxidación, ha demostrado que es necesario para la germinación conidial, por lo que se hizo una revisión en cuanto a las rutas metabólicas de las enzimas y cómo estás pueden también de manera indirecta afectar su recurso energético en el hongo Magnaporthe grisea. (Rajesh N et al., 2012) Esto por un lado muestra lo importante de los compuestos que se encuentran en la naturaleza, para intentar atacar a C.acutatum a nivel mitocondrial.

 Por otro lado existen también antecedentes con otros organismos que han atacado por la acción de los lipopéptidos, en alguno de ellos por la amalgama de las 3 familias y en otras con lipopéptidos bien definidos en cuanto a su familia. Por ejemplo en el estudio realizado por Arroyave-Toro et al. (2017), muestran como el biocontrol por estos compuestos como los lipopéptidos son de gran importancia en cuando a las enfermedades postcosecha.

 Algo interesante es que se evaluó el sobre nadante libre de células que quedaba luego de la obtención de los metabolitos secundarios para mirar que capacidad de inhibición tenía y arrojó que si existía inhibición, por tanto luego de un acercamiento mejor a los compuestos dio como resultado que las iturinas A y las fengicinas C inhibían a C.acutatum con concentraciones mínimas como lo son a 32 ppm, como sea quiere decir que todos estos compuestos tienen en menor o mayor medida una capacidad inhibitoria contra este fitopatógeno por ejemplo. 

No quedándose ahí, también se ha probado en otros fitopatógenos como lo es Ralstonia solanacearum, en el estudio de Villegas-Escobar et al. (2018), donde siguiendo el mismo proceso que en los anteriores con la identificación de los compuestos, en este caso lipopéptidos, se tiene una inhibición de este patógeno, demostrando una vez más la acción inhibitoria que puede tener, si bien este estudio se realiza en una bacteria, el mecanismo por el cual ataca la misma y como ataca el hongo, van de la mano, en cuanto a que permeabilizan la membrana y produce el desequilibrio interno.


CONCLUSIÓN

En conclusión el alineamiento de secuencias parece provechoso en la medida que nos muestra que organismos tienen similitud en sus clusters de genes o en un gen específico para producir un metabolito de interés en este caso se evaluaron estas dos bacterias y se obtuvo que son muy parecidas en cuanto a esta región por lo que podrían ser estudiadas desde la parte experimental para conocer que cantidad de estos lipopéptidos produce y si realmente si puede inhibir un fitopatógeno como C.acutatum
