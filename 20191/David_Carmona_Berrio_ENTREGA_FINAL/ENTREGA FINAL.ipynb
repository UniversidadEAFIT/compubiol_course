# Introduction

<div class="text-justify">The count of the usage frequency for many codons respect to the amino acids that are synthesized during the translation results interesting when evaluating for the genetic code that diverse organisms use to their own amino acids synthesis, this phenomenon is known as Codon usage bias where every organism has a preference to use sort kind of codons instead of the genetic code established as a general way; the molecular evolutionist researchers suggest that this phenomenon is manifested as a result of the balance between the mutational and translational selection of the genes, also this phenomenon is very widely between species and can contribute to the genome evolution in a significant way (Behura & Severson, 2013).</div><br>
    

<div class="text-justify">There are 64 codons to the synthesis of the amino acids (3 codons to stop and 1 to begin the translation process that is the methionine), so really there are 61 codons available to the synthesis of 20 amino acids. So we have to keep in mind that there are different codons that synthesize to the same amino acid, those codons are called “sinonimum codons” and it has been demonstrated that some organisms use different codons alternative to the sinonimum codons in a preferential way to synthesize some amino acid, in other words, this is not an aleatory process, mentioned before as usage codon bias (Behura & Severson, 2013). The importance of the study of codon usage bias is mainly in the information we get about one organism that can provide us information as researchers about the main function of sort kind of gene, also about the evolutive process related; it is even valuable because of the expression of heterologous genes as a powerful tool in biotechnological processes that obtains recombinant products as insulin, antibiotics, and vaccines for example. This codon usage bias of the host organism results in a limiting factor to the product obtentions and would be more useful the codon optimization to really get the recombinant product we are looking for, doing this would not affect the expression of the recombinant gene in the host (Behura & Severson, 2013). Whereby using this bioinformatic tool where you can observe the production of amino acids by analyzing the sequence that produces it and would be useful to quantify the frequency of the amino acids that depends on which genetic code will be using booth organisms.</div><br>


<div class="text-justify"><i>E. coli</i> has been demonstrated that the preference of codon usage to the start codon is ATG, 3542; GTG, 612; TTG, 130; respectively (Blattner et al. 1997). Having this pattern on mind we can now imagine how <i>E. coli</i> works to express diverse proteins due to their codon preference, also, the expression of functional proteins in heterologous hosts is a limiting factor of the modern biotechnology (mentioned earlier), because of the difficult to express the proteins outside their original context (Gustafsson, Govindarajan & Minshull, 2004). It is particular of the codon usage bias that every organism can use to express it in their context, the improvements in the speed and cost of gene synthesis have facilitated the complete redesign of entire gene sequences to maximize the likelihood of high desired protein expression (Gustafsson, Govindarajan & Minshull, 2004). </div><br>

<div class="text-justify">The protein expression is easily affected by the codon usage bias and the reason it could be being affected is by the preferred codons correlated with the abundance of tRNAs available within the cell. This relationship serves to optimize the translational system and to balance codon concentration with isoacceptor tRNA concentration during translation (Gustafsson, Govindarajan & Minshull, 2004). In <i>E. coli</i>, for example, the arginine tRNA that reads the infrequently used AGG and AGA codons to synthesize arginine is present only at very low levels, so it is likely that codon usage and isoacceptor tRNA concentrations have coevolved, and that the selection pressure for this co-evolution is more pronounced for highly expressed genes than for genes expressed at low levels (Gustafsson, Govindarajan & Minshull, 2004). Realizing then the complications that result in the expression of sort kind of proteins and even more in heterologous proteins, would be interesting to solve this problem trying to visualize which genetic code prefers the organisms that are going to be an object of cloning, everything to visualize better how could be compatible between them, looking for codons usage preference.</div>

# Methodology

<div class="text-justify">This work is based on Python code, using the numpy, matplotlib and Bio libraries. After having all the libraries available and running on Phyton you can start to digitate the code getting both sequences from the organisms that will be an object of cloning, in this case, a sequence from Variovorax sp. EA-ED0101 C306 that has not been reported on the GenBank, nonetheless there is <i>E. coli</i> Blue-mar on the other site, that will be the heterologous organism to clone the gene that synthesizes the hypothetical protein from Variovorax sp. EA-ED0101. When the sequences are available and downloaded from the GenBank then you can start to design your code, and have to keep in mind that the sequences you are using have to be encode to amino acids, and everything respect to the RNA will be deleted because there is no translation process and can affect the conformation of the whole triplets, because later we need to group it to split it, so the RNA that are not in triplets will disrupt the normal sequence and be an error factor. Once you have everything codifying you can group the sequences and split it by triplets, then you can group it again looking for those codons who do not repeat, after grouping it you can count it on the triplet group you split before in the whole sequence who many times does this triplet is repeated. Having this group of counting codons, then you can calculate their frequency and later graph it with numply and matplotlib help; now you can compare it between the organisms and see which codons preferences do they have. Finally, as a general commentary respectely to the sequences sizes, Variovorax sp. EA-ED0101 is 34kpb and <i>E. coli</i> is 150kpb long.</div>

```python
def Secuencias(nombre_archivo):
    
    from Bio import SeqIO
    handle = open(nombre_archivo)
    global secuenciapr
    seq_full=[]
    for seq_record in SeqIO.parse(handle, "fasta"):
        #print(seq_record.id)
        seq_full.append(str(seq_record.seq))
        #print(len(seq_record))
        
    handle.close()
#    print(seq_full[30]) Esta secuencia se tiene que eliminar porque hace referencia a un tRNA-Met(cat)
    secuenciapr=''

# # # Como estoy trabajando con una secuencia tan grande, quiero corroborar que me esté agrupando todas en la lista de string
#    contador=0
#    i=0
#    while i<=len(seq_full)-1:
#        contador=contador+len(seqfull[i]) 
#        i=i+1
# # # Hasta aquí va el contador, luego lo imprimes donde quieras ver.
    
    i=0
    while i<=29:
        secuenciapr=secuenciapr+seq_full[i] #Aquí se está agrupando todas secuencias en un string desde la 1 hasta la 30
        i=i+1
    
    i=31
    while i<=len(seq_full)-1:
        secuenciapr=secuenciapr+seq_full[i] #Aquí se está agrupando todas secuencias en un string desde la 31 hasta la 38 con el fin de omitir la 31 que no es codificante para proteína
        i=i+1
        
    print(len(secuenciapr)) #Secuencia completa sin el gen 30 de tRNA
    return secuenciapr
 
Secuencias("PROKKA_05272019_150K.ffn")
```

```python
def ConteoCodones(seq):
    
    global recuento_glob
    global conteo_glob
    global frecuencias_glob
    global enumero_glob
    
    tripleta=[]
    ini=0
    fin=3
    while ini<=len(seq)-1:
        tripleta.append(seq[ini:fin])
        fin=fin+3
        ini=ini+3
    print(tripleta); print()
    print(len(tripleta));print() #Corroborar el número de tripletas que tengo
    
    recuento=[]
    i=0
    while i<=len(tripleta)-1:
        if tripleta[i] not in recuento:
            recuento.append(tripleta[i]) #Aquí meto en una lista los codones que existen (que no se repitan)
        i=i+1
    print(recuento)
    recuento_glob=recuento
    
    # Ya que tengo separados la cantidad de codones y los codones que no se repiten, puedo consultar cuantas veces se repiten
    
    conteo=[]
    i=0
    while i<=len(recuento)-1:
        conteo.append(tripleta.count(recuento[i]))
        i=i+1
    print()    
    print(conteo);print();print()
    conteo_glob=conteo
    
    # Ahora voy a tratar de realizar un diccionario con cada TRIPLETA (recuento) y su FRECUENCIA (conteo)
    
    dicci={}
    i=0
    while i<=len(recuento)-1:
        dicci[recuento[i]]=conteo[i]
        i=i+1
        
    # Se procede a realizar las frecuencias sobre el número total de estas secuencias que es 11372
    
    print(dicci)
    
    frecuencias=[]
    i=0
    while i<=len(conteo)-1:
        frecuencias.append(((conteo[i]/len(tripleta))*100))
        i=i+1
        
    #Conteo de lista
    
    enumero=[]
    i=1
    while i<=len(frecuencias):
        enumero.append(i)
        i=i+1
    
    enumero_glob=enumero
    print()
    print(enumero)
    print()
    print(frecuencias)
    frecuencias_glob=frecuencias
    
    
    
ConteoCodones(secuenciapr)
```

```python
def grafica(nose):

    import numpy as np
    import matplotlib.mlab as mlab
    import matplotlib.pyplot as plt


    # example data
    x = recuento_glob
    y = frecuencias_glob
    
    plt.figure(figsize=(27, 9))
    plt.bar(x, y)
    plt.xlabel('Codons', fontsize="20")
    plt.ylabel('Frequency', fontsize="20")
    plt.ylim(0,10)
    plt.title(r'Count of codons frequency in Escherichia coli', fontsize="30")
    
    plt.show()

    
grafica("holi")
```

# Analysis and conclusions

<div class="text-justify">Once you can viasualize both graphs (Count of codons frequency in Variovorax sp. EA-ED0101 and count of codons frequency in <i>E. coli</i>) you can watch explicitly their codon preference usage to the synthesis of amino acids they are going to use the most, however that does not tell us anything explicitly only give us a picture of their codon usage bias, but this information is not worthless because it also give us the global panorama about how much is viable to realize a heterologous expression of some protein that depends on which start codon their use on their gene and how much composition does use the organism to evaluate posteriorly the viability of the heterologous expression.</div><br>

<div class="text-justify">So, one of the resulting problems here is particularly the codon usage bias, where <i>E. coli</i> (the model organism) is always the object of cloning to obtain the heterologous gene, and it is important to talk about the hungry codon syndrome, where <i>E. coli</i> specifically presents a tRNA population closely matched to the overall codon bias of the resident mRNA population (Kurland & Gallant, 1996). Cloned heterologous genes often differ markedly in codon bias, and, even in the crude relative amino acid content of their encoded proteins (Kurland & Gallant, 1996).</div><br>

<div class="text-justify">An excellent example of the mentioned case before is the Insuline, that has a cysteine content much higher tan that of total <i>E. coli protein</i> (expressed as heterologous protein). Within codon families, <i>E. coli</i> avoids certain codons that mammalian genes use more abundantly. Therefore, the high-level expression of a cloned heterologous gene may place demands on the host protein synthetic that are not matched to its normal tRNA population; as a result, ribosomes programmed with the heterologous message may stop at positions calling for a tRNA that is either rare itself, present at low concentrations or largely deacylated because of the heavier than normal drain of its amino acid into protein (Kurland & Gallant, 1996). So, maybe the protein synthesis may be successful in heterologous organisms but there are many gaps on the real protein synthesis because of the nature of the different organisms that are going to be cloned (as donor and host).</div><br>

<div class="text-justify">Knowing this would be interesting to design a strategy to increase the efficiency of the process to really obtain the wished protein, by optimizing heterologous protein expression is more effective in this process. There are features of a robust prokaryotic expression system that discuss a variety of strategies to optimize protein production in <i>E. coli.</i> And some strategies are going to be mentioned briefly as the components that are essential of the expression vectors, which are the promoters because of the capability that has of enhance the cellular production of sort kind of compounds, the transcription termitators, that are placed upstream of the promoter that drives expression of the gene of interest and will minimize background transcription because an excessive expression will destabilize the plasmid that contains the gene of interest, the elements affecting the translational process which unique structural features at the 5’ end of the mRNA transcript are the major determinants of the efficiency of mRNA translational initiation, the translational enhancers, sequence elements from <i>E. coli</i> and bacteriophages have been shown to enhance heterologous gene expression in E. coli markedly, the translational terminator, that usually <i>E. coli</i> uses three stop codons consecutevely to prevent ribosome “skipping”, the mRNA stabilizers, the codon usage and finally the choice of cellular compartment for protein expression because it also affects the expression (Hanning & Markrides, 1998). So there are many strategies that can be applied to the process.</div><br>

<div class="text-justify">In conclusion, there are so many codon usage bias between organisms that can rely on evolutionary processes and more, but for heterologous expression results complicated because of the cloning, and for many divergence between codon usage in different organisms so this tool is applied to see that divergence and prefence by the organism, as you read before there are some complications that can not be corrected in the heterologous expression but in the codon optimization process you can make this less likelihood to syntesize the amino acids that we are not looking for in the protein synthesis into the heterologous expression.</div>
